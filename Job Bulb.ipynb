```python
# importing the necessary libraries
import pint
u = pint.UnitRegistry()
import numpy as np
import matplotlib.pyplot as plt
```

```python
# measured properties of the Job bulbs F3-F 68C (Red- 20mm)
thickness = 248.76 / 732.463 * u.millimeter
print("Thickness of the glass tube: " , thickness)

# measured mass of bulb with liquid
m_total = 1.4473 * u.gram / 5.0

# mass of bulb without the liquid
m_glass = 0.9287 * u.gram / 5.0
print("Mass of the glass : ", m_glass)

# mass of the liquid
m_liquid = m_total - m_glass
print("Mass of the liquid : ", m_liquid)
print("Total mass of the F3-F bulb : ", m_total)
print("Total mass of the dual activation bulb : ", 0.2343*u.gram)
print("m_glass/m_liquid = ", m_glass/m_liquid)

# Calculating the amount of energy needed to bring the bulb to its activation temperature.
c_glass = 0.84 * u.joule / u.gram / u.kelvin
c_liquid = 181.2 * u.joule / u.mole / u.kelvin * (1.0/92.14) * u.mole / u.gram # liquid = Toluene
c_silver = 0.233 * u.joule / u.gram / u.kelvin
T_initial = (273.2 + 20.0) * u.kelvin
T_activation = (273.2 + 68.0) * u.kelvin
Q_glass = m_glass * c_glass * (T_activation - T_initial)
Q_liquid = m_liquid * c_liquid * (T_activation - T_initial)
Q_total = Q_glass + Q_liquid
print("Q_glass :", Q_glass)
print("Q_liquid: ", Q_liquid)
print("Total energy : ", Q_total)
print("Q_glass/Q_liquid = ", Q_glass/Q_liquid)

mass_glass = {'F3-F 68C 20mm': m_glass.magnitude}
mass_fluid = {'F3-F 68C 20mm': m_liquid.magnitude}
heat_fluid = {'F3-F 68C 20mm': Q_liquid.magnitude}
heat_glass = {'F3-F 68C 20mm': Q_glass.magnitude}
heat_total = {'F3-F 68C 20mm': Q_total.magnitude}
mass_total = {'F3-F 68C 20mm': m_total.magnitude}
heat_ratio = {'F3-F 68C 20mm': Q_glass.magnitude/Q_liquid.magnitude}
mass_ratio = {'F3-F 68C 20mm': m_glass.magnitude/m_liquid.magnitude}
heat_liquid_total = {'F3-F 68C 20mm': Q_liquid.magnitude/Q_total.magnitude}
```

```python
# measured properties of the Job bulbs Straight 3mm 68C (Red- 20mm)
thickness = 248.76 / 732.463 * u.millimeter
print("Thickness of the glass tube: " , thickness)

# measured mass of bulb with liquid
m_total = 0.4535 * u.gram / 2.0

# mass of bulb without the liquid
m_glass = 0.8721 * u.gram / 5.0
print("Mass of the glass : ", m_glass)

# mass of the liquid
m_liquid = m_total - m_glass
print("Mass of the liquid : ", m_liquid)
print("Total mass of the bulb : ", m_total)
print("Total mass of the dual activation bulb : ", 0.2343*u.gram)
print("m_glass/m_liquid = ", m_glass/m_liquid)

# mass of the metalic trace
m_trace = (0.2339+0.2353+0.2335+0.2342+0.2340)/5.0 *u.gram

# Calculating the amount of energy needed to bring the bulb to its activation temperature.
c_glass = 0.84 * u.joule / u.gram / u.kelvin
c_liquid = 181.2 * u.joule / u.mole / u.kelvin * (1.0/92.14) * u.mole / u.gram # liquid = Toluene
T_initial = (273.2 + 20.0) * u.kelvin
T_activation = (273.2 + 68.0) * u.kelvin
Q_glass = m_glass * c_glass * (T_activation - T_initial)
Q_liquid = m_liquid * c_liquid * (T_activation - T_initial)
Q_total = Q_glass + Q_liquid
print("Q_glass :", Q_glass)
print("Q_liquid: ", Q_liquid)
print("Total energy : ", Q_total)
print("Q_glass/Q_liquid = ", Q_glass/Q_liquid)
print("m_trace = ", m_trace)
Q_trace = m_trace * c_silver * (T_activation - T_initial)
print("Q_with_trace = ", Q_total + Q_trace)

mass_glass['Straight 68C 20mm'] = m_glass.magnitude
mass_fluid['Straight 68C 20mm'] = m_liquid.magnitude
heat_fluid['Straight 68C 20mm'] = Q_liquid.magnitude
heat_glass['Straight 68C 20mm'] = Q_glass.magnitude
heat_total['Straight 68C 20mm'] = Q_total.magnitude
mass_total['Straight 68C 20mm'] = m_total.magnitude
heat_ratio['Straight 68C 20mm'] = Q_glass.magnitude/Q_liquid.magnitude
mass_ratio['Straight 68C 20mm'] = m_glass.magnitude/m_liquid.magnitude
heat_liquid_total['Straight 68C 20mm'] = Q_liquid.magnitude/Q_total.magnitude
```

```python
# Total heat and time needed to bring a F3-F 68C 20mm to its operating temperature 

# measured mass of bulb with liquid
m_total = 1.4473 * u.gram / 5.0

# mass of bulb without the liquid
m_glass = 0.9287 * u.gram / 5.0

# mass of the liquid
m_liquid = m_total - m_glass

# Calculating the amount of energy needed to bring the bulb to its activation temperature.
c_glass = 0.84 * u.joule / u.gram / u.kelvin
c_liquid = 181.2 * u.joule / u.mole / u.kelvin * (1.0/92.14) * u.mole / u.gram # liquid = Toluene
c_silver = 0.233 * u.joule / u.gram / u.kelvin
T_initial = (273.2 + 20.0) * u.kelvin
T_activation = (273.2 + 68.0) * u.kelvin
Q_glass = m_glass * c_glass * (T_activation - T_initial)
Q_liquid = m_liquid * c_liquid * (T_activation - T_initial)
Q_total = Q_glass + Q_liquid

Temperature = (np.linspace(-30,68,1000)+273.2)*u.kelvin
Q=np.zeros(Temperature.size)*u.joule
Q = m_glass * c_glass * (T_activation - Temperature)+m_liquid * c_liquid * (T_activation - Temperature)
plt.rcParams['figure.figsize'] = (18, 5)
plt.rcParams.update({'font.size': 12})
#tick_label does the some work as plt.xticks()
plt.subplot(1,2,1)
plt.plot(Temperature-273.2*u.kelvin, Q)
plt.ylabel('Energy (Joule)')
plt.xlabel('Initial Temperature ($^\circ C$)')
plt.title('Energy needed to bring the bulb to the operating temperature of 68$^\circ C$')
plt.grid()

plt.subplot(1,2,2)
Qpt = Q/(5.6494*u.watt)
plt.plot(Temperature-273.2*u.kelvin,Qpt)
plt.ylabel('Time (s)')
plt.xlabel('Initial Temperature ($^\circ C$)')
plt.title('Time needed to bring the bulb to the operating temperature of 68$^\circ C$')
plt.grid()
plt.show()
```
